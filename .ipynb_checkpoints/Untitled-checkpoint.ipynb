```python

import pandas as pd
import tensorflow as tf
import tensorflow_hub as hub
# import tensorflow_text as text
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df_spam = df[df['Category'] == 'spam']
df_spam.shape
```

```python
df_ham = df[df['Category'] == 'ham']
df_ham.shape
```

```python
df_ham_downshaped = df_ham.sample(df_spam.shape[0])
df_ham_downshaped.shape
```

```python
df_balanced = pd.concat([df_spam,df_ham_downshaped ])
```
